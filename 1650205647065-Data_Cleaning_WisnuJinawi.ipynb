```python
df = df.rename(columns={'customerID':'CustID'})
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df['paymentmethod']
```

```python
df['paymentmethod'] = df['paymentmethod'].str.replace('automatic','')
df['paymentmethod']
```

```python
df['paymentmethod'] = df['paymentmethod'].str.replace('()','', regex = False)
df['paymentmethod']
```

```python
df['paymentmethod'] = df['paymentmethod'].str.strip()
df['paymentmethod']
```

```python
df.head()
```

```python
#one hor encoding

ohe = pd.get_dummies(df['phoneservice'],prefix = 'service')
ohe
```

```python
df = df.join(ohe)
df
```

```python
#normalization
df['monthlycharges_norm'] = MinMaxScaler().fit_transform(df['monthlycharges'].values.reshape(len(df),1))
df.head()
```

```python
#standardization (sebelumnya ada problem)
df['totalcharges_standard'] = StandardScaler().fit_transform(df['totalcharges'].values.reshape(len(df),1))
df.head()
```
